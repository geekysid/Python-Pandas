```python
import pandas as pd
file = 'Data/Survey.csv'
df = pd.read_csv(file, index_col="Respondent", 
                 usecols=["Respondent", "Hobbyist", "Country", 
                          "Student", "YearsCode", "ConvertedComp", 
                          "LanguageWorkedWith", "Age", "Gender"])
df.head()
```

```python
df.columns
```

```python
df.columns = ['Column-1', 'Column-2', 'Column-3', 'Column-4', 
                  'Column-5', 'Column-6', 'Column-7', 'Column-8']
```

```python
df.columns
```

```python
df.head()
```

```python
df.columns = df.columns.str.replace('-', '_')
df.columns
```

```python
df.columns = df.columns.str.lower()
df.columns
```

```python
# Usecase of rename() - Changing name of column, 'ConvertedComp', to 'SalaryInUsd'
df.rename(columns = {'ConvertedComp' : 'SalaryInUsd'}, inplace = True)
df.columns
```

```python
df.head()
```

```python
# Changing Gender of Respondent 1 to Female
df.loc[1, 'Gender'] = 'Female'
df.head()
```

```python
# Changing Gender of Respondent 2, 3, and 4 to Female
df.loc[[2, 3, 4], 'Gender'] = 'Female'
df.head()
```

```python
# # Using Conditionals to Change Multiple Row

# Before Changing Man to Male
man_condition = df['Gender'] == 'Man'
male_condition = df['Gender'] == 'Male'
man_count = df.loc[man_condition, 'Gender'].count()
male_count = df.loc[male_condition, 'Gender'].count()
print("Before Changing Man to Male")
print(f"Rows with Gender as Man: {man_count}")
print(f"Rows with Gender as Male: {male_count}")

# Changing gender of all rows with gender as Man to Male
df.loc[man_condition, 'Gender'] = 'Male'
print()

# After Changing Man to Male
man_condition = df['Gender'] == 'Man'
male_condition = df['Gender'] == 'Male'
man_count = df.loc[man_condition, 'Gender'].count()
male_count = df.loc[male_condition, 'Gender'].count()
print("After Changing Man to Male")
print(f"Rows with Gender as Man: {man_count}")
print(f"Rows with Gender as Male: {male_count}")
```
