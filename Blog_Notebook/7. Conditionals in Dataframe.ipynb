```python
import pandas as pd
file = 'Data/Survey.csv'
df = pd.read_csv(file, index_col="Respondent", 
                 usecols=["Respondent", "Hobbyist", "Country", "Student", "YearsCode",
                          "ConvertedComp", "LanguageWorkedWith", "Age", "Gender"])
df.head(15)
```

```python
# (column) (locgical-operator) (value)
# condition: Respondents who do coding as Hobby
condition = (df["Hobbyist"] == "Yes")
print("Type: ", type(condition))
print()
print("Series: \n", condition)
```

```python
# df.condition
# df[condition]  # this is simillar to df[(df["Hobbyist"] == "Yes")]
df.loc[condition]  # this is simillar to df.loc[(df["Hobbyist"] == "Yes")]
```

```python
# Respondent from India
condition2 = (df["Country"] == "India")

# Respondent who code as Hobby and are from India
condition3 = (condition & condition2)
# the above code is equivalent to 
# condition3 = (df["Hobbyist"] == "Yes") & (df["Country"] == "India")
```

```python
df.loc[condition3]
```
