```python
import pandas as pd
```

```python
df = pd.read_csv('survey_results_public.csv')  # impproting survey_results_public csv file and creating a dataframe
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.columns.to_list() # list of all columns in dataframe
```

```python
# creating a filter that has all rows as true that has 'Hobbyist' column as 'Yes' otherwise as False.
fltr = (df['Hobbyist'] == 'Yes')
```

```python
# using the filter to output only rows that has rows as True or that satify the condition
df[fltr].head()
```

```python
# we can also use loc to fetch the results using fltr
df.loc[fltr].head()
```

```python
# the benifit of using loc or iloc is that we can also select the columns that we want to ouput. 
# like in below example, we are outputting 3 columns, 'Respondent', 'Hobbyist' and 'ConvertedComp'
df.loc[fltr, ['Respondent', 'Hobbyist', 'Salary']].head()
```

```python
# We can also use AND (&) and OR (|) operator to filter for more then one condition. 
# Filter that sets rows to True if respondent has Yes as hobbyist and salary of over 100,000 usd
fltr = ((df['Hobbyist'] == 'Yes') & (df['Salary'] > 100000))
```

```python
# using the filter in our dataframe with loc
df.loc[fltr, ['Respondent', 'Hobbyist', 'Salary']]
```

```python
# we can also get the count of rows that satisfy the condition
df.loc[fltr, 'Respondent'].count()
```

```python
# Using NOT operator 
# we can use ~ (tilda) sign to get the opposite of the filter.
df.loc[~fltr, 'Respondent'].count()
```

```python
# Finding rows where Respondents have not shared salary
fltr = (df['Salary'].isnull())
df.loc[fltr, ['Respondent', 'Hobbyist', 'Salary']].head()
```

```python
# df['CloumnName'].isnull() : returns rows as True who has a column as NaN.
fltr = df['Hobbyist'].isnull()
df.loc[fltr, 'Respondent'].count() # we have 0 number of Respondent with NaN as their answer for the Hobbyist question
```

```python
# we have 60 number of Respondent with NaN as their answer for the Salary question
fltr = df['Salary'].isnull()
df.loc[fltr, 'Respondent'].count()
```

```python
# df['CloumnName'].notnull() : This is opposite of isnull() It returns rows as true 
# if the value of the column is not NaN
fltr = df['Salary'].notnull()
```

```python
# we have 140 number of Respondent with valid answer for the Salary question as n
df.loc[fltr, 'Respondent'].count()
```

```python
# Excercise: Finding to 5 most common languages that people have worked with.

from collections import Counter  # import Counter

fltr = df['LanguageWorkedWith'].notnull()  # filter to remove NaN from our datasframe.
language = df.loc[fltr, 'LanguageWorkedWith'].to_list()  # applying filter and getting data in the form of list 
c_lang_known = Counter()  # creating a counter that counts the languages

for item in language:  # looping through all the items in language list
    lang_list = item.split(';')  # spliting each item in a list after spliting with ; to create another list
    c_lang_known.update(lang_list)  # using counter update method to count each element in the list

print(c_lang_known)  # printing the outcome
c_lang_known.most_common(5)  # printing the top 5 most common language
```
