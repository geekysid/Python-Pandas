```python
# counting number of occurance of unique value in a Series
hobbyist_count = df['Hobbyist'].value_counts()
hobbyist_count
```

```python
type(hobbyist_count)
```

```python
hobbyist_count_dict = hobbyist_count.to_dict() # converting Series to dictionary
hobbyist_count_dict
```

```python
for item in hobbyist_count_dict:
    print(f'{item}: {hobbyist_count_dict[item]}')
```

```python
# iloc: we use iloc to fetch data from dataframe using integer index 
# even if index of rows and columns are in strings

print(df.iloc[1]) # returns 2nd row from the dataframe
print()
print(df.iloc[1, 2]) # returns 2nd row and 3rd(Hobbyist) coloumn from the dataframe
```

```python
# iloc cont.
df.iloc[[0,1,2,3], [2,3,5]] # fetching rows with index 0,1,2 and 3 & columns with index 2,3,4 and 5 from dataframe
```

```python
# iloc cont.
# using slice to fech rows and column. It is important to note that end index is not a part of outpart
df.iloc[0:3, 2:5]
```

```python
# loc: we use iloc to fetch data from dataframe using actual labels of rows and column
print(df.loc[1]) # returns 2nd row from the dataframe. Here actual index of rows is integer
print()
df.loc[1, 'Hobbyist'] # returns 2nd row and Hobbyist coloumn from the dataframe. Here actual index of rows is integer and that of column is string
```

```python
# loc cont. 
# In this dataframe, actual index of rows is integer and that of column is string
df.loc[[0,1,2,3], ['Hobbyist', 'OpenSourcer', 'Employment']] # fetching rows with index 0,1,2 and 3 & columns  with index 2, 3 and 5 from dataframe
```

```python
# loc cont.
# In this dataframe, actual index of rows is integer and that of column is string
# using slice to fech rows and column.
df.loc[0:3, 'Hobbyist': 'Employment'] # fetching rows with index 0,1,2 and 3 & columns with index 2,3,4 and 5 from dataframe
```
