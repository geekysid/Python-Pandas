```python
# fetching max, min, std, mean for a particular column with int or float datatype
max_profit = df['Total Profit'].max()
print('max_profit: ' + str(max_profit))
min_profit = df['Total Profit'].min()
print('min_profit: ' + str(min_profit))
mean_profit = df['Total Profit'].mean()
print('mean_profit: ' + str(mean_profit))
std_profit = df['Total Profit'].std()
print('std_profit: ' + str(std_profit))
```
