```python
import pandas as pd

# creating Dataframe from csv files
df = pd.read_csv('SalesRecords_Sample.csv')
df.head()
```

```python
# creating Dataframe from excel files
df = pd.read_excel('SalesRecords_Sample.xlsx', header=1)  # header=1 specifies positin of header in the file
df.head()
```

```python
# creating dataframe from list of tuple
lst_of_tpl = [(1, 2, 3, 4, 5, 6), ('a', 'b', 'c', 'd', 'e', 'f'), ('g', 'h', 'i', 'j', 'k', 'l')]
df = pd.DataFrame(lst_of_tpl, columns=[1,2,3,4,5,6])
df
```

```python
# Row Oriented: from list of dictionary with each dict acting as a row and keys are column name 
# and values are vlaue of the column
list_dict = [{'Name': 'Manish', 'Age': 32, 'Salary': '$2.5K'}, 
             {'Name': 'Latish', 'Age': 30, 'Salary': '$3.0K'}, 
             {'Name': 'Satish', 'Age': 25, 'Salary': '$3.5K'}]
df = pd.DataFrame(list_dict)
df
```

```python
# Column Oriented: from dictionary with keys acting as column name and  values(list) acting as 
# all the values of the columns
list_dict = {'Name': ['Manish', 'Latish', 'Satish'],
              'Age': [32, 30, 25],
              'Salary': ['$2.5K', '$3.5K', '$3.5K']}
df = pd.DataFrame(list_dict)
df
```
